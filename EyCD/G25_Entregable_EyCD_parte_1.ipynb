Para filtrar los valores extremos utilizamos el método de los 3 sigmas: filtramos los valores por encima o por debajo de 3 desvios estandar de la media.

```python
melb_df = melbourne_df.copy()
```

```python
def filter_outliers_mask(df, cols, n_desvios_ = 3, verbosity = True):

  df_size = df.shape[0]

  idx_to_delete=[]

  for c in cols:
    print(c)
    media = df[c].mean()
    desvio = df[c].std()
    print(desvio/media)
    
    minimo, maximo = (media - n_desvios * desvio, media + n_desvios * desvio)
    
    mask = (df[c] < minimo) | (df[c] > maximo)

    idx_to_delete.append(list(df.loc[mask].index))
    
  idx_to_delete = numpy.array([item for sublist in idx_to_delete for item in sublist])

  if verbosity:
    print("DF Size", df_size)
    print(f"Eliminaremos un {numpy.round(len(idx_to_delete)/df_size * 100,1)}% de los registros")
    print("DF Processed size: ", df_size - len(idx_to_delete))

  
  return idx_to_delete

def filter_df(df, idx_to_delete):
  return df.drop(index = idx_to_delete)



def filter_categorical(df, cols, perc_of_cut = 0.005, verbosity = True):  
  df_size = df.shape[0]
  idx_to_delete=[]  

  for c in categorical_cols:   
    categories_count = df[c].value_counts()/df[c].count()
    categories_to_delete = categories_count[categories_count < perc_of_cut].index
    idx_to_delete_categorical = df[df[c].isin(categories_to_delete)].index
    idx_to_delete.append(idx_to_delete_categorical)
 
 
  idx_to_delete = numpy.array([item for sublist in idx_to_delete for item in sublist])

  if verbosity:
    print("DF Size", df_size)
    print(f"Eliminaremos un {numpy.round(len(idx_to_delete)/df_size * 100,1)}% de los registros")
    print("DF Processed size: ", df_size - len(idx_to_delete))

  
  return idx_to_delete
```

```python
categorical_cols = ['Type',"Suburb",'CouncilArea']
numerical_cols = ['Rooms',  'Bedroom2', 'Bathroom','Distance',  'Car',
       'Landsize', 'BuildingArea', 'YearBuilt', 'Price']
n_desvios = 3

# Numerical filtering 3 sigma
print("Numerical\n")
idx_to_delete_numerical = filter_outliers_mask(melb_df, numerical_cols, n_desvios_ = n_desvios, verbosity = True)

# Categorical filtering
print("\n Categorical")
idx_to_delete_categorical = filter_categorical(melb_df, categorical_cols, perc_of_cut = 0.001, verbosity = True)

idx_to_delete = numpy.append(idx_to_delete_numerical, idx_to_delete_categorical)
idx_to_delete = numpy.unique(idx_to_delete)
```

## Ejercicio 3:

Crear y guardar un nuevo conjunto de datos con todas las transformaciones realizadas anteriormente.

```python
# Read Melb dataframe
melb_df = pandas.read_csv(
    'https://cs.famaf.unc.edu.ar/~mteruel/datasets/diplodatos/melb_data.csv')
# Read airbnb dataframe
airbnb_df = pandas.read_csv(
    'https://cs.famaf.unc.edu.ar/~mteruel/datasets/diplodatos/cleansed_listings_dec18.csv')

# Drop rows with outliers using the 3 sigma rule and frequency in categorical variables
categorical_cols = ['Type',"Suburb",'CouncilArea']
numerical_cols = ['Rooms',  'Bedroom2', 'Bathroom','Distance',  'Car',
       'Landsize', 'BuildingArea', 'YearBuilt','Price']

cols = numerical_cols + categorical_cols + ["Postcode"]

n_desvios = 3


# Numerical filtering 3 sigma
print("Numerical\n")
idx_to_delete_numerical = filter_outliers_mask(melb_df, numerical_cols, n_desvios_ = n_desvios, verbosity = True)

# Categorical filtering
print("\n Categorical")
idx_to_delete_categorical = filter_categorical(melb_df, categorical_cols, perc_of_cut = 0.001, verbosity = True)

idx_to_delete = numpy.append(idx_to_delete_numerical, idx_to_delete_categorical)
idx_to_delete = numpy.unique(idx_to_delete)

# Melbourne DataFrame
melb_df = filter_df(melb_df, idx_to_delete)
melb_df = melb_df[cols]


# Get Zipcode's features from airbnb dataset
relevant_cols = ['zipcode','neighborhood', 'price', 'weekly_price', 'monthly_price']

# Groupby
airbnb_price_by_zipcode = airbnb_df[relevant_cols].groupby('zipcode')\
  .agg({'price': ['min', 'mean', "max"], 'weekly_price': ['min', 'mean', "max"],
        'monthly_price': ['min', 'mean', "max"], "zipcode": "count"})\
  .reset_index()
# Flatten the two level columns
airbnb_price_by_zipcode.columns = [
  ' '.join(col).strip()
  for col in airbnb_price_by_zipcode.columns.values]
# Rename columns
airbnb_price_by_zipcode = airbnb_price_by_zipcode.rename(
    columns={'price min': "airbnb_price_min",
             'price mean': 'airbnb_price_mean',
             'price max':'airbnb_price_max',
             'weekly_price min': 'airbnb_weekly_price_min',
             'weekly_price mean': 'airbnb_weekly_price_mean',
             'weekly_price max': 'airbnb_weekly_price_max',
             'monthly_price min': 'airbnb_monthly_price_min',
             'monthly_price mean': 'airbnb_monthly_price_mean',
             'monthly_price max': 'airbnb_monthly_price_max',
             'zipcode count': 'zipcode_count',
             'neighborhood': 'airbnb_neighborhood'
                }
)

# Filter  
n_registros = 25
airbnb_price_by_zipcode_min= airbnb_price_by_zipcode[airbnb_price_by_zipcode["zipcode_count"] >= n_registros] 

# Change type
airbnb_price_by_zipcode_min.loc[:, "zipcode"] = airbnb_price_by_zipcode_min.zipcode.astype(float)
melb_df.loc[:, "Postcode"] = melb_df.Postcode.astype(float)

# Merge
merged_sales_df = melb_df.merge(
    airbnb_price_by_zipcode_min, how='left',
    left_on='Postcode', right_on='zipcode'
)

merged_sales_df.to_csv("/content/melb_data_processed.csv", index=False)
```
